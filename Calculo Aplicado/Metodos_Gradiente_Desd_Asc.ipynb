```python
# Método de gradiente descendente

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Definir la función
def f(x, y):
    return x**2 + y**2 + 5 * np.sin(x) + 10 * np.cos(y)

# Definir el gradiente
def gradient(x, y):
    df_dx = 2 * x + 10 * np.cos(x)
    df_dy = 2 * y - 10 * np.sin(y)
    return np.array([df_dx, df_dy])

# Gradiente descendente (mínimos)
def gradient_descent(start, learning_rate, max_iterations, tolerance):
    x = np.array(start, dtype=float)
    trajectory = [x.copy()]
    for i in range(max_iterations):
        grad = gradient(x[0], x[1])
        if np.linalg.norm(grad) < tolerance:
            print(f"Convergencia en iteración {i}")
            break
        x = x - learning_rate * grad
        trajectory.append(x.copy())
    return x, f(x[0], x[1]), np.array(trajectory)

# Gradiente ascendente (máximos)
def gradient_ascent(start, learning_rate, max_iterations, tolerance):
    x = np.array(start, dtype=float)
    trajectory = [x.copy()]
    for i in range(max_iterations):
        grad = gradient(x[0], x[1])
        if np.linalg.norm(grad) < tolerance:
            print(f"Convergencia en iteración {i}")
            break
        x = x + learning_rate * grad
        trajectory.append(x.copy())
    return x, f(x[0], x[1]), np.array(trajectory)

# Parámetros
start_point = [0.0, 0.0]
learning_rate = 0.01
max_iterations = 1000
tolerance = 1e-5

# Ejecutar algoritmos
min_point, min_value, min_trajectory = gradient_descent(start_point, learning_rate, max_iterations, tolerance)
max_point, max_value, max_trajectory = gradient_ascent(start_point, learning_rate, max_iterations, tolerance)

# Imprimir resultados
print(f"Mínimo en: {min_point}, Valor: {min_value}")
print(f"Máximo en: {max_point}, Valor: {max_value}")

# Visualización: Gráfico de contornos con trayectorias
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

plt.figure(figsize=(8, 6))
plt.contour(X, Y, Z, levels=20, cmap='viridis')
plt.plot(min_trajectory[:, 0], min_trajectory[:, 1], 'b.-', label='Descenso (Mínimo)')
plt.plot(max_trajectory[:, 0], max_trajectory[:, 1], 'r.-', label='Ascenso (Máximo)')
plt.scatter([min_point[0]], [min_point[1]], c='blue', s=100, label='Mínimo')
plt.scatter([max_point[0]], [max_point[1]], c='red', s=100, label='Máximo')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Trayectorias del Gradiente Descendente y Ascendente')
plt.legend()
plt.savefig('trajectories.png')
plt.show()

# Visualización: Superficie 3D
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.8)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x, y)')
ax.set_title('Función: x² + y² + 10sin(x) + 10cos(y)')
plt.savefig('surface.png')
plt.show()
```

```python
# Método de gradiente descendente.

import numpy as np
import matplotlib.pyplot as plt


# Definir la función
def f(x):
    return (x**2 * np.cos(x) - x) / 10

# Definir el gradiente en 1D
def gradiente(x):
    # Derivada
    df_dx = (2*x*np.cos(x) - x**2*np.sin(x) - 1) / 10
    return df_dx

# Gradiente descendente para función
def gradiente_descendente(inicio, tasa_aprendizaje, max_iteraciones, tolerancia):
    x = float(inicio)  # Convertimos a float para problema 1D
    trayectoria = [x]
    for i in range(max_iteraciones):
        grad = gradiente(x)
        if abs(grad) < tolerancia:
            print(f"Convergencia alcanzada en iteración {i}")
            break
        x = x - tasa_aprendizaje * grad #En la opuesta dirección del gradiente
        trayectoria.append(x)
    return x, f(x), np.array(trayectoria)

# Parámetros
punto_inicial = 3.0 # Punto inicial (1D)
tasa_aprendizaje = 0.05 # Tasa de aprendizaje reducida para mejor convergencia
max_iteraciones = 1000
tolerancia = 1e-5

# Ejecutamos los pasos del algoritmo
min_punto, min_valor, min_trayectoria = gradiente_descendente(
    punto_inicial, tasa_aprendizaje, max_iteraciones, tolerancia
)

# Imprimir
print(f"Mínimo encontrado en x = {min_punto:.6f}")
print(f"Valor de la función en el mínimo: {min_valor:.6f}")

# Visualización en 1D
x_vals = np.linspace(-5, 5, 400)
y_vals = f(x_vals)

plt.figure(figsize = (10, 6))
plt.plot(x_vals, y_vals, 'b-', label = 'Función f(x)')
plt.plot(min_trayectoria, f(min_trayectoria), 'ro-',
         markersize=4, label='Trayectoria del gradiente')
plt.scatter([min_punto], [min_valor], c ='green', s = 100,
            label = f'Mínimo encontrado ({min_punto:.2f}, {min_valor:.2f})')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Método de gradiente descendente para f(x) = (x²·cos(x) - x)/10')
plt.legend()
plt.grid(True)
plt.savefig('gradiente_descendente_1d.png')
plt.show()

plt.savefig('/content/gradiente_ascendente_1d.png')
#Convergencia alcanzada en iteración 131
#Mínimo encontrado en x = 3.701260
#Valor de la función en el mínimo: -1.531050
```
